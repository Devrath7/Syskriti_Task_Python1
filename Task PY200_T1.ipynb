```python
#MEAN
numbers = [10, 20, 36, 40, 50]

total = 0
count = 0

for num in numbers:
    total = total + num
    count = count + 1

if count != 0:
    mean = total / count
    print("Mean is:", mean)
else:
    print("List is empty. Cannot compute mean.")
```

```python
##MEDIAN
numbers = [7, 2, 9, 4, 5]

# Step 1: Manually calculate length
count = 0
for _ in numbers:
    count = count + 1

# Step 2: Manual Sorting (Bubble Sort)
i = 0
while i < count - 1:
    j = 0
    while j < count - i - 1:
        if numbers[j] > numbers[j + 1]:
            temp = numbers[j]
            numbers[j] = numbers[j + 1]
            numbers[j + 1] = temp
        j = j + 1
    i = i + 1

# Step 3: Calculate Median
if count % 2 == 1:
    middle_index = count // 2
    median = numbers[middle_index]
else:
    middle_index1 = (count // 2) - 1
    middle_index2 = count // 2
    median = (numbers[middle_index1] + numbers[middle_index2]) / 2

print("Sorted List:", numbers)
print("Median is:", median)
```

```python
###MODE
numbers = [4, 2, 2, 3, 4, 4, 5, 2, 4]

length = 0
for _ in numbers:
    length = length + 1

max_count = 0
mode = None

i = 0
while i < length:
    current = numbers[i]
    count = 0

    j = 0
    while j < length:
        if numbers[j] == current:
            count = count + 1
        j = j + 1

    if count > max_count:
        max_count = count
        mode = current

    i = i + 1

print("Mode is:", mode)
print("Frequency is:", max_count)
```
