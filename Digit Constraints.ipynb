```python
# Find smallest number with given digit constraints

N = int(input("Enter N: "))

result = -1

for x in range(1, N + 1):
    s = str(x)
    
    # Condition 1: No digit 0
    if '0' in s:
        continue
    
    # Condition 2: Exactly two 3s
    if s.count('3') != 2:
        continue
    
    # Condition 3: Sum of digits divisible by 7
    digit_sum = sum(int(d) for d in s)
    if digit_sum % 7 != 0:
        continue
    
    result = x
    break  # since we need smallest

print(result)
```
